```python
def line_from_points(x1,y1,x2,y2):
    m=(y2-y1)/(x2-x1)
    c=y1-m*x1
    def mx_plus_c(x):
        return m*x+c
    return mx_plus_c

def puva_erythema(wavelength):
    eq=line_from_points(325,1.0,387,0)
    if wavelength < 325.0:
        pv=1.0
    elif wavelength < 387:
        pv=eq(wavelength)
    else:
        pv=0
    return pv

puv_ery=np.vectorize(puva_erythema)
```
